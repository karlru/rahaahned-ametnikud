```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Private average wage

df = pd.read_csv('data/keskmine_brutopalk_sektor.csv')

df_2018 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
private_average_2018 = df_2018.iloc[3]

df_2019 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
private_average_2019 = df_2019.iloc[3]

df_2020 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
private_average_2020 = df_2020.iloc[3]

df_2021 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
private_average_2021 = df_2021.iloc[3]

print("Private average 2018 (bruto): " + private_average_2018 + "€ \n" + 
      "Private average 2019 (bruto): " + private_average_2019 + "€ \n" + 
      "Private average 2020 (bruto): " + private_average_2020 +"€ \n" + 
      "Private average 2021 (bruto): " + private_average_2021 + "€")
```

```python
data_private = [[2018,private_average_2018], [2019, private_average_2019], [2020,private_average_2020],[2021, private_average_2021]]
df_private = pd.DataFrame(data_private, columns=['Year', 'Average private sector wage (€)'])
df_private['Average private sector wage (€)']=df_private['Average private sector wage (€)'].astype(float)
print(df_private)
private_plot = df_private.plot(x='Year', y='Average private sector wage (€)', xticks=(2018, 2019, 2020, 2021), grid=True,figsize=(10, 7),title='Average wage in Private sector 2018-2021')
```

```python
# State average wage

d_2018_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
state_average_2018 = d_2018_riik.iloc[1]

d_2019_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
state_average_2019 = d_2019_riik.iloc[1]

d_2020_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
state_average_2020 = d_2020_riik.iloc[1]

d_2021_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
state_average_2021 = d_2021_riik.iloc[1]

print("Private (just State) average 2018 (bruto): " + state_average_2018 + "€ \n" + 
      "Private (just State) average 2019 (bruto): " + state_average_2019 + "€ \n" + 
      "Private (just State) average 2020 (bruto): " + state_average_2020 +"€ \n" + 
      "Private (just State) average 2021 (bruto): " + state_average_2021 + "€")
```

```python
data_state = [[2018,state_average_2018], [2019, state_average_2019], [2020,state_average_2020],[2021, state_average_2021]]
df_state = pd.DataFrame(data_state, columns=['Year', 'Average state wage (€)'])
df_state['Average state wage (€)']=df_state['Average state wage (€)'].astype(float)
print(df_state)
state_plot = df_state.plot(x='Year', y='Average state wage (€)',xticks=(2018, 2019, 2020, 2021), grid=True, title='Average wage in State institutions 2018-2021')
```

```python
# Municipality Government average wage

d_2018_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
municipality_government_average_2018 = d_2018_kov.iloc[2]

d_2019_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
municipality_government_average_2019 = d_2019_kov.iloc[2]

d_2020_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
municipality_government_average_2020 = d_2020_kov.iloc[2]

d_2021_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
municipality_government_average_2021 = d_2021_kov.iloc[2]

print("Private (just Municipality Government) average 2018 (bruto): " + municipality_government_average_2018 + "€ \n" + 
      "Private (just Municipality Government) average 2019 (bruto): " + municipality_government_average_2019 + "€ \n" + 
      "Private (just Municipality Government) average 2020 (bruto): " + municipality_government_average_2020 +"€ \n" + 
      "Private (just Municipality Government) average 2021 (bruto): " + municipality_government_average_2021 + "€")
```

```python
data_municipality_government = [[2018, municipality_government_average_2018], [2019, municipality_government_average_2019], [2020,municipality_government_average_2020],[2021, municipality_government_average_2021]]
df_municipality_government = pd.DataFrame(data_municipality_government, columns=['Year', 'Average Municipality Government wage (€)'])
df_municipality_government['Average Municipality Government wage (€)']=df_municipality_government['Average Municipality Government wage (€)'].astype(float)
print(df_municipality_government)
municipality_government_plot = df_municipality_government.plot(x='Year', y='Average Municipality Government wage (€)',xticks=(2018, 2019, 2020, 2021), grid=True, title='Average wage in Municipality Government 2018-2021')
```

```python
#Public Sector average wage 


df_ministers_2018 = pd.read_csv('data/pohipalk_2018.csv')
average_2018 = round(df_ministers_2018['Põhipalk'].sum()/df_ministers_2018['Põhipalk'].count(),0)

df_ministers_2019 = pd.read_csv('data/pohipalk_2019.csv')
average_2019 = round(df_ministers_2019['Põhipalk'].sum()/df_ministers_2019['Põhipalk'].count(),0)

df_ministers_2020 = pd.read_csv('data/pohipalk_2020.csv')
average_2020 = round(df_ministers_2020['Põhipalk'].sum()/df_ministers_2020['Põhipalk'].count(),0)

df_ministers_2021 = pd.read_csv('data/pohipalk_2021.csv')
average_2021 = round(df_ministers_2021['Põhipalk'].sum()/df_ministers_2021['Põhipalk'].count(),0)

print("Public average 2018 (bruto): " + str(average_2018) + "€ \n" + 
      "Public average 2019 (bruto): " + str(average_2019) + "€ \n" + 
      "Public average 2020 (bruto): " + str(average_2020) +"€ \n" + 
      "Public average 2021 (bruto): " + str(average_2021) + "€")
```

```python
data_public_average = [[2018, average_2018], [2019, average_2019], [2020, average_2020],[2021, average_2021]]
df_public_average = pd.DataFrame(data_public_average, columns=['Year', 'Average public sector wage (€)'])
print(df_public_average)
plot = df_public_average.plot(x='Year', y='Average public sector wage (€)', xticks=(2018, 2019, 2020, 2021) , grid=True,
                              title='Average wage in Public Sector 2018-2021')
```
