```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {}
for year in ['2018', '2019', '2020', '2021']:
    data[year] = pd.read_csv(f'data/pohipalk_{year}.csv')
    
    if year in ['2018', '2019', '2020']:
        data[year] = data[year].iloc[:,4:]
    else:
        data[year] = data[year].iloc[:,2:]
        
    data[year]['Palk täiskohaga'] = data[year]['Põhipalk'] / data[year]['Ametniku koormus ametkohal']
```

```python
institution_averages_2018 = data['2018'].groupby(by=['Asutus']).mean(True)
institution_averages_2018.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk', 'Asutuse keskmine palk täiskohaga']
institution_gender_averages_2018 = data['2018'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2018.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk', 'Keskmine palk täiskohaga']

institution_averages_2019 = data['2019'].groupby(by=['Asutus']).mean(True)
institution_averages_2019.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk', 'Asutuse keskmine palk täiskohaga']
institution_gender_averages_2019 = data['2019'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2019.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk', 'Keskmine palk täiskohaga']

institution_averages_2020 = data['2020'].groupby(by=['Asutus']).mean(True)
institution_averages_2020.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk', 'Asutuse keskmine palk täiskohaga']
institution_gender_averages_2020 = data['2020'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2020.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk', 'Keskmine palk täiskohaga']

institution_averages_2021 = data['2021'].groupby(by=['Asutus']).mean(True)
institution_averages_2021.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk', 'Asutuse keskmine palk täiskohaga']
institution_gender_averages_2021 = data['2021'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2021.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk', 'Keskmine palk täiskohaga']
```

```python
joined_averages_2018 = institution_gender_averages_2018.join(institution_averages_2018, on="Asutus").sort_values(by=['Asutuse keskmine palk täiskohaga', 'Asutus'])
joined_averages_2019 = institution_gender_averages_2019.join(institution_averages_2019, on="Asutus").sort_values(by=['Asutuse keskmine palk täiskohaga', 'Asutus'])
joined_averages_2020 = institution_gender_averages_2020.join(institution_averages_2020, on="Asutus").sort_values(by=['Asutuse keskmine palk täiskohaga', 'Asutus'])
joined_averages_2021 = institution_gender_averages_2021.join(institution_averages_2021, on="Asutus").sort_values(by=['Asutuse keskmine palk täiskohaga', 'Asutus'])
```
