```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {}
for year in ['2018', '2019', '2020', '2021']:
    data[year] = pd.read_csv(f'data/pohipalk_{year}.csv')
    
    if year in ['2018', '2019', '2020']:
        data[year] = data[year].iloc[:,4:]
    else:
        data[year] = data[year].iloc[:,2:]
```

```python
institution_averages_2018 = data['2018'].groupby(by=['Asutus']).mean(True)
institution_averages_2018.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk']
institution_gender_averages_2018 = data['2018'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2018.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk']

institution_averages_2019 = data['2019'].groupby(by=['Asutus']).mean(True)
institution_averages_2019.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk']
institution_gender_averages_2019 = data['2019'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2019.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk']

institution_averages_2020 = data['2020'].groupby(by=['Asutus']).mean(True)
institution_averages_2020.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk']
institution_gender_averages_2020 = data['2020'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2020.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk']

institution_averages_2021 = data['2021'].groupby(by=['Asutus']).mean(True)
institution_averages_2021.columns = ['Asutuse keskmine koormus ametkohal', 'Asutuse keskmine põhipalk']
institution_gender_averages_2021 = data['2021'].groupby(by=['Asutus', 'Sugu']).mean(numeric_only=True)
institution_gender_averages_2021.columns = ['Keskmine koormus ametkohal', 'Keskmine põhipalk']
```

```python
institution_gender_averages_2018 = institution_gender_averages_2018.reset_index(level=['Sugu'])
institution_gender_averages_2018_N = institution_gender_averages_2018[institution_gender_averages_2018['Sugu'] == 'N']
institution_gender_averages_2018_M = institution_gender_averages_2018[institution_gender_averages_2018['Sugu'] == 'M']
institution_gender_averages_2018 = institution_gender_averages_2018_N.join(institution_gender_averages_2018_M, lsuffix=' N', rsuffix=' M')
institution_gender_averages_2018 = institution_gender_averages_2018[['Keskmine koormus ametkohal N', 'Keskmine põhipalk N', 'Keskmine koormus ametkohal M', 'Keskmine põhipalk M']]

institution_gender_averages_2019 = institution_gender_averages_2019.reset_index(level=['Sugu'])
institution_gender_averages_2019_N = institution_gender_averages_2019[institution_gender_averages_2019['Sugu'] == 'N']
institution_gender_averages_2019_M = institution_gender_averages_2019[institution_gender_averages_2019['Sugu'] == 'M']
institution_gender_averages_2019 = institution_gender_averages_2019_N.join(institution_gender_averages_2019_M, lsuffix=' N', rsuffix=' M')
institution_gender_averages_2019 = institution_gender_averages_2019[['Keskmine koormus ametkohal N', 'Keskmine põhipalk N', 'Keskmine koormus ametkohal M', 'Keskmine põhipalk M']]

institution_gender_averages_2020 = institution_gender_averages_2020.reset_index(level=['Sugu'])
institution_gender_averages_2020_N = institution_gender_averages_2020[institution_gender_averages_2020['Sugu'] == 'N']
institution_gender_averages_2020_M = institution_gender_averages_2020[institution_gender_averages_2020['Sugu'] == 'M']
institution_gender_averages_2020 = institution_gender_averages_2020_N.join(institution_gender_averages_2020_M, lsuffix=' N', rsuffix=' M')
institution_gender_averages_2020 = institution_gender_averages_2020[['Keskmine koormus ametkohal N', 'Keskmine põhipalk N', 'Keskmine koormus ametkohal M', 'Keskmine põhipalk M']]

institution_gender_averages_2021 = institution_gender_averages_2021.reset_index(level=['Sugu'])
institution_gender_averages_2021_N = institution_gender_averages_2021[institution_gender_averages_2021['Sugu'] == 'N']
institution_gender_averages_2021_M = institution_gender_averages_2021[institution_gender_averages_2021['Sugu'] == 'M']
institution_gender_averages_2021 = institution_gender_averages_2021_N.join(institution_gender_averages_2021_M, lsuffix=' N', rsuffix=' M')
institution_gender_averages_2021 = institution_gender_averages_2021[['Keskmine koormus ametkohal N', 'Keskmine põhipalk N', 'Keskmine koormus ametkohal M', 'Keskmine põhipalk M']]
institution_gender_averages_2021.head()
```

```python
joined_averages_2018 = institution_averages_2018.join(institution_gender_averages_2018)
joined_averages_2019 = institution_averages_2019.join(institution_gender_averages_2019)
joined_averages_2020 = institution_averages_2020.join(institution_gender_averages_2020)
joined_averages_2021 = institution_averages_2021.join(institution_gender_averages_2021)
```

```python
joined_averages_2018
```

```python
joined_df = joined_averages_2018.join(joined_averages_2019, lsuffix=' 2018', rsuffix=' 2019')
joined_df = joined_df.join(joined_averages_2020, rsuffix=' 2020')
joined_df = joined_df.join(joined_averages_2021, rsuffix=' 2021')
joined_df = joined_df[['Asutuse keskmine põhipalk 2018', 'Keskmine põhipalk N 2018', 'Keskmine põhipalk M 2019', 'Asutuse keskmine põhipalk 2019', 'Keskmine põhipalk N 2019', 'Keskmine põhipalk M 2019', 'Asutuse keskmine põhipalk', 'Keskmine põhipalk N', 'Keskmine põhipalk M', 'Asutuse keskmine põhipalk 2021', 'Keskmine põhipalk N 2021', 'Keskmine põhipalk M 2021']]
joined_df.columns = ['Asutus keskmine 2018', 'N keskmine 2018', 'M keskmine 2018', 'Asutus keskmine 2019', 'N keskmine 2019', 'M keskmine 2019', 'Asutus keskmine 2020', 'N keskmine 2020', 'M keskmine 2020', 'Asutus keskmine 2021', 'N keskmine 2021', 'M keskmine 2021']
joined_df
```

```python
joined_df['Lõhe 2018'] = (joined_df['M keskmine 2018'] - joined_df['N keskmine 2018']) / joined_df['M keskmine 2018'] * 100
joined_df['Lõhe 2019'] = (joined_df['M keskmine 2019'] - joined_df['N keskmine 2019']) / joined_df['M keskmine 2019'] * 100
joined_df['Lõhe 2020'] = (joined_df['M keskmine 2020'] - joined_df['N keskmine 2020']) / joined_df['M keskmine 2020'] * 100
joined_df['Lõhe 2021'] = (joined_df['M keskmine 2021'] - joined_df['N keskmine 2021']) / joined_df['M keskmine 2021'] * 100
joined_df['Lõhe keskmine'] = joined_df[['Lõhe 2018', 'Lõhe 2019', 'Lõhe 2020', 'Lõhe 2021']].mean(axis=1)
joined_df.head()
```

```python
joined_df[['Lõhe 2018', 'Lõhe 2019', 'Lõhe 2020', 'Lõhe 2021']].describe()
```

```python
joined_df = joined_df.sort_values(by=['Lõhe keskmine'])
joined_df.head()
```

```python
highest_gaps = joined_df[-14:-4]
highest_gaps
```

```python
joined_df['Tõus 2019'] = joined_df['Asutus keskmine 2019'] * 100 / joined_df['Asutus keskmine 2018'] - 100
joined_df['Tõus 2020'] = joined_df['Asutus keskmine 2020'] * 100 / joined_df['Asutus keskmine 2019'] - 100
joined_df['Tõus 2021'] = joined_df['Asutus keskmine 2021'] * 100 / joined_df['Asutus keskmine 2020'] - 100
joined_df['Tõus keskmine'] = joined_df[['Tõus 2019', 'Tõus 2020', 'Tõus 2021']].mean(axis=1)
joined_df.head()
```

```python
joined_df[['Tõus 2019', 'Tõus 2020', 'Tõus 2021', 'Tõus keskmine']].describe()
```

```python
joined_df = joined_df.sort_values(by=['Tõus keskmine'])
joined_df.head()
```

```python
highest_growers = joined_df[:-3]
highest_growers
```

```python
x1 = ['2018', '2019', '2020', '2021']
y1 = [joined_df['Lõhe 2018'].mean(axis=0), joined_df['Lõhe 2019'].mean(axis=0), joined_df['Lõhe 2020'].mean(axis=0), joined_df['Lõhe 2021'].mean(axis=0)]
x2 = ['2019', '2020', '2021']
y2 = [joined_df['Tõus 2019'].mean(axis=0), joined_df['Tõus 2020'].mean(axis=0), joined_df['Tõus 2021'].mean(axis=0)]

plt.plot(x1, y1, label='Average gender pay gap')
plt.plot(x2, y2, label='Average pay raise compared to previous year')

plt.legend()
plt.show()
```

```python
lohed = joined_df.dropna(axis=0, subset=['Lõhe 2018', 'Lõhe 2019', 'Lõhe 2020', 'Lõhe 2021'])
lohed = lohed[['Lõhe 2018', 'Lõhe 2019', 'Lõhe 2020', 'Lõhe 2021']]
lohed.columns = ['2018', '2019', '2020', '2021']
ax = lohed.T.plot(color='#D3D3D3')

keskmised_lohed = [18.0, 17.1, 15.6, 14.9]
avg, = ax.plot(lohed.columns, keskmised_lohed, label='Average gender pay gap', color='b')
for x_value, y_value in zip(avg.get_xdata(), avg.get_ydata()):
    label = "{:.2f}".format(y_value)
    ax.annotate(label,(x_value, y_value), xytext=(0, 5), textcoords="offset points", ha='center', va='bottom')

ax.legend([avg, ax.get_lines()[0]], ['Average gender pay gap', 'Various public institutions'])

plt.title('Comparison of gender pay gap in public institutions compared to the general average', fontsize=8)

plt.savefig('data/paygaps.png', dpi=300, bbox_inches='tight', transparent=True)
```
