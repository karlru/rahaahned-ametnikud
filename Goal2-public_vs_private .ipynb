```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# siin era keskmised
# on 'Eesti eraõiguslik isik' 3. rida

df = pd.read_csv('data/keskmine_brutopalk_sektor.csv')

df_2018 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
private_average_2018 = df_2018.iloc[3]

df_2019 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
private_average_2019 = df_2019.iloc[3]

df_2020 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
private_average_2020 = df_2020.iloc[3]

df_2021 = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
private_average_2021 = df_2021.iloc[3]

print("Private average 2018 (bruto): " + private_average_2018 + "€ \n" + 
      "Private average 2019 (bruto): " + private_average_2019 + "€ \n" + 
      "Private average 2020 (bruto): " + private_average_2020 +"€ \n" + 
      "Private average 2021 (bruto): " + private_average_2021 + "€")
```

```python
data_private = [[2018,private_average_2018], [2019, private_average_2019], [2020,private_average_2020],[2021, private_average_2021]]
df_private = pd.DataFrame(data_private, columns=['Year', 'Average wage in private sector 2018-2021'])
df_private['Average wage in private sector 2018-2021']=df_private['Average wage in private sector 2018-2021'].astype(float)
print(df_private)
private_plot = df_private.plot(x='Year', y='Average wage in private sector 2018-2021', xticks=(2018, 2019, 2020, 2021), grid=True)
```

```python
# siin riik keskmised
# 'Riik' on 1. rida

df_2018_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
state_average_2018 = df_2018_riik.iloc[1]

df_2019_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
state_average_2019 = df_2019_riik.iloc[1]

df_2020_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
state_average_2020 = df_2020_riik.iloc[1]

df_2021_riik = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
state_average_2021 = df_2021_riik.iloc[1]

print("Private (just State) average 2018 (bruto): " + state_average_2018 + "€ \n" + 
      "Private (just State) average 2019 (bruto): " + state_average_2019 + "€ \n" + 
      "Private (just State) average 2020 (bruto): " + state_average_2020 +"€ \n" + 
      "Private (just State) average 2021 (bruto): " + state_average_2021 + "€")
```

```python
data_state = [[2018,state_average_2018], [2019, state_average_2019], [2020,state_average_2020],[2021, state_average_2021]]
df_state = pd.DataFrame(data_state, columns=['Year', 'Average state wage 2018-2021'])
df_state['Average state wage 2018-2021']=df_state['Average state wage 2018-2021'].astype(float)
print(df_state)
state_plot = df_state.plot(x='Year', y='Average state wage 2018-2021',xticks=(2018, 2019, 2020, 2021), grid=True)
```

```python
# siin kov keskmised
# 'Kov' on 2. rida

df_2018_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2018"]
municipality_government_average_2018 = df_2018_kov.iloc[2]

df_2019_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2019"]
municipality_government_average_2019 = df_2019_kov.iloc[2]

df_2020_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2020"]
municipality_government_average_2020 = df_2020_kov.iloc[2]

df_2021_kov = pd.read_csv('data/keskmine_brutopalk_sektor.csv',header = 1).loc[:,"2021"]
municipality_government_average_2021 = df_2021_kov.iloc[2]

print("Private (just Municipality Government) average 2018 (bruto): " + municipality_government_average_2018 + "€ \n" + 
      "Private (just Municipality Government) average 2019 (bruto): " + municipality_government_average_2019 + "€ \n" + 
      "Private (just Municipality Government) average 2020 (bruto): " + municipality_government_average_2020 +"€ \n" + 
      "Private (just Municipality Government) average 2021 (bruto): " + municipality_government_average_2021 + "€")
```

```python
data_municipality_government = [[2018, municipality_government_average_2018], [2019, municipality_government_average_2019], [2020,municipality_government_average_2020],[2021, municipality_government_average_2021]]
df_municipality_government = pd.DataFrame(data_municipality_government, columns=['Year', 'Average Municipality Government wage 2018-2021'])
df_municipality_government['Average Municipality Government wage 2018-2021']=df_municipality_government['Average Municipality Government wage 2018-2021'].astype(float)
print(df_municipality_government)
municipality_government_plot = df_municipality_government.plot(x='Year', y='Average Municipality Government wage 2018-2021',xticks=(2018, 2019, 2020, 2021), grid=True)
```

```python
# siin avaliksektor kokku  

public_average_2018 = int(state_average_2018) + int(municipality_government_average_2018)
public_average_2019 = int(state_average_2019) + int(municipality_government_average_2019)
public_average_2020 = int(state_average_2020) + int(municipality_government_average_2020)
public_average_2021 = int(state_average_2021) + int(municipality_government_average_2021)
print("Public average 2018 (bruto): " + str(public_average_2018) + "€ \n" + 
      "Public average 2019 (bruto): " + str(public_average_2019) + "€ \n" + 
      "Public average 2020 (bruto): " + str(public_average_2020) +"€ \n" + 
      "Public average 2021 (bruto): " + str(public_average_2021) + "€")
```

```python
data_public_average = [[2018, public_average_2018], [2019, public_average_2019], [2020, public_average_2020],[2021, public_average_2021]]
df_public_average = pd.DataFrame(data_public_average, columns=['Year', 'Average public sector wage'])
df_public_average['Average public sector wage']=df_public_average['Average public sector wage'].astype(float)
print(df_public_average)
plot = df_public_average.plot(x='Year', y='Average public sector wage', xticks=(2018, 2019, 2020, 2021) , grid=True)
```

```python
#Kõik ühte dataframe

df_all = pd.DataFrame()

df_all = pd.merge(pd.merge(df_municipality_government, df_state, on='Year'),
                 df_private, on='Year').merge(df_public_average, on='Year')
df_all.head()

yticks=np.arange(1000,3400,step=100)
yticks2=np.arange(0,3400,step=250)


plot_all=df_all.plot(x='Year', y=['Average Municipality Government wage 2018-2021',
                         'Average state wage 2018-2021', 
                         'Average wage in private sector 2018-2021', 'Average public sector wage'], 
            xticks=(2018, 2019, 2020, 2021), grid=True, figsize=(10, 7)).set_yticks(yticks)


plot_all2 = df_all.plot.bar(x='Year', y=['Average Municipality Government wage 2018-2021',
                         'Average state wage 2018-2021', 
                         'Average wage in private sector 2018-2021', 'Average public sector wage'], figsize=(10, 7), grid = True).set_yticks(yticks2)


```
