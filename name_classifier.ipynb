```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
def classifyGender(name):
    response = requests.get(f'https://www.stat.ee/nimed/{name}')
    soup = BeautifulSoup(response.content, 'html.parser')
    info_field = soup.find('section').find('p').getText()

    male = re.search('(\d*) (?:mehel)', info_field)
    male = (int(male[1]) if male[1] else 0) if male is not None else 0

    female = re.search('(\d*) (?:naisel)', info_field)
    female = (int(female[1]) if female[1] else 0) if female is not None else 0

    gender = 'M' if male > female else 'N'
    return gender
```

```python
years = ['2018', '2019', '2020', '2021']
cache = {}
```

```python
def classifyGenderWithCache(name):
    if name in cache:
        return cache[name]
        
    gender = classifyGender(name)
    cache[name] = gender
    return gender

for year in years:
    data = pd.read_csv(f'data/pohipalk_{year}.csv')
    data['Sugu'] = 'T'
    rows = data.shape[0]
    
    data['Sugu'] = data.apply(lambda row: classifyGenderWithCache(row['Eesnimi']), axis=1)
    
    data.to_csv(f'data/pohipalk_{year}.csv')
```
